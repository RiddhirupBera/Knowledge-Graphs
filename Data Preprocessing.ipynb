# Dataset

### Stock price dataset

```python
stock_price = pd.read_csv('/home/riddhirup/Downloads/GOOG.csv')
```

```python
stock_price.head()
```

```python
stock_price = stock_price.drop(columns=['Volume','KDJ','W%R','RSI'])
```

### Stock news dataset

```python
stock_news = pd.read_csv('/home/riddhirup/Downloads/google.csv')
```

```python
stock_news.head()
```

```python
stock_news = stock_news.groupby(['Date'])['Title'].apply(', '.join).reset_index()
```

```python
stock_news.head()
```

### Merged dataset

```python
merged_data = pd.merge(stock_news,stock_price,how="right",on="Date")
```

```python
merged_data.head()
```

# Preprocessing

```python
merged_data.info()
```

### Assigning the labels

#### Labels:
####    0: price decreases or stayed the same
####    1: price increased

```python
merged_data["Avg_Price"]=(merged_data["Open"]+merged_data["Close"])/2
```

```python
merged_data.head(10)
```

```python
merged_data["Percentage_Change"] = merged_data["Avg_Price"].pct_change()
```

```python
merged_data["Percentage_Change"] = merged_data["Percentage_Change"].shift(periods=-1)
```

```python
merged_data = merged_data.dropna()
```

```python
merged_data
```

```python
merged_data["Label"] = merged_data.Percentage_Change.apply(lambda x:1 if x>0 else 0)
```

```python
merged_data
```

```python
merged_data.info()
```

```python
merged_data["Label"].value_counts()
```

```python
merged_data = merged_data.drop(columns = ["Open","High","Low","Close","Avg_Price","Percentage_Change"])
```

```python
merged_data.to_csv("Google_dataset.csv")
```
